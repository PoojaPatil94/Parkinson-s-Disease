```python
import h2o
```

```python
from h2o.automl import H2OAutoML
import pandas as pd
```

```python
h2o.init()
```

```python
df = h2o.import_file('H:\Spring_2020\AI\project\parkinsons_train.csv')
```

```python
df_test = h2o.import_file('H:\Spring_2020\AI\project\parkinsons_test.csv')
```

```python
df_test.head()
```

```python
train, test = df.split_frame(ratios = [.7])
```

```python
x = train.columns
y = "status"
x.remove(y)
```

```python
#https://www.kaggle.com/ateplyuk/titanic-automl


# For binary classification, response should be a factor
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
```

```python
# Run AutoML for 900 seconds
aml = H2OAutoML(max_runtime_secs = 900)
aml.train(x = x, y = y, training_frame = train, leaderboard_frame = test)
```

```python
aml.leaderboard
aml.leader
```

```python
preds = aml.leader.predict(df_test)
```

```python
pr = preds['predict']
df_pr = pr.as_data_frame()
```

```python
result = df_test['name'].as_data_frame()
result.insert(1,'PD', df_pr)
result.to_csv('Result.csv', index=False)
```
